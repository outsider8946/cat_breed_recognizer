```python
import os
import pandas as pd
```

```python
PATH_TO_DATA = '/content/drive/MyDrive/Colab Notebooks/cat_breed_recognizer/data/'
cat_breeds = ['bengal', 'maine_coon', 'ragdoll', 'oriental_shorthair', 'british_shorthair', 'siamese']
label_map = {'bengal':0, 'maine_coon':1, 'ragdoll':2, 'oriental_shorthair':3, 'british_shorthair':4, 'siamese':5}
```

```python
df = pd.read_csv(PATH_TO_DATA+cat_breeds[-1]+f'/{cat_breeds[-1]}.csv')
df
```

```python
for breed in cat_breeds:
  if os.path.exists(PATH_TO_DATA+breed+f'/{breed}.csv'):
    df = pd.read_csv(PATH_TO_DATA+breed+f'/{breed}.csv',on_bad_lines='skip')
    for i in range(len(df)):
      width  = df.loc[i,'image_width']
      height = df.loc[i,'image_height']
      df.loc[i,'bbox_x'] = df.loc[i,'bbox_x']/width
      df.loc[i,'bbox_y'] = df.loc[i,'bbox_y']/height
      df.loc[i,'bbox_width'] = df.loc[i,'bbox_width']/width
      df.loc[i,'bbox_height'] = df.loc[i,'bbox_height']/height
    df['label_name'] = label_map[df.loc[0,'label_name']]
    df = df.drop(['image_width','image_height'],axis = 1)
    df = df.rename(columns = {'bbox_x':'norm_bbox_x','bbox_y':'norm_bbox_y', 'bbox_width':'norm_bbox_width','bbox_height':'norm_bbox_height'})
    df.to_csv(PATH_TO_DATA+breed+f'/{breed}.csv',index=False)
```
